```python
# Импортируем необходимые библиотеки
import plotly.graph_objects as go
from plotly.subplots import make_subplots

df = pd.DataFrame(data)

# Преобразуем timestamp в формат datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Фильтруем данные по каждому активу
zerobro_df = df[df['symbol'] == 'ZEREBROUSDT']
ada_df = df[df['symbol'] == 'ADAUSDT']
popcat_df = df[df['symbol'] == 'POPCATUSDT']

# Функция для построения графика
def plot_candlestick_with_volume(data, title):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.1,
                        subplot_titles=(title, 'Volume'), row_width=[0.2, 0.7])

    # График свечей
    fig.add_trace(go.Candlestick(
        x=data['timestamp'],
        open=data['open'],
        high=data['high'],
        low=data['low'],
        close=data['close'],
        name='Candlesticks'
    ), row=1, col=1)

    # Столбчатый график объёмов
    fig.add_trace(go.Bar(
        x=data['timestamp'],
        y=data['volume'],
        name='Volume'
    ), row=2, col=1)

    # Настройки макета
    # fig.update_layout(
    #     title=title,
    #     xaxis_title='Time',
    #     yaxis_title='Price',
    #     xaxis_rangeslider_visible=False
    # )
    fig = make_subplots(
        rows=3,
        cols=1,
        shared_xaxes=True,
        vertical_spacing=0.1,  # Расстояние между графиками
        subplot_titles=(title, 'Volume'),
        row_width=[0.2, 0.2, 0.6],  # Относительная высота строк
        row_heights=[0.6, 0.2, 0.2]  # Новый параметр: задает высоту каждой строки
    )
    fig.show()

# Построим графики для каждого актива
plot_candlestick_with_volume(zerobro_df, 'ZEREBROUSDT Candlestick Chart')
plot_candlestick_with_volume(ada_df, 'ADAUSDT Candlestick Chart')
plot_candlestick_with_volume(popcat_df, 'POPCATUSDT Candlestick Chart')
```

# Построение графиков Bollinger Bands и ATR

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def plot_candlestick_with_indicators(data, title, window=14, multiplier=2):
    """
    Функция для построения графика свечей с объемами, Bollinger Bands и ATR.

    Параметры:
    - data: DataFrame с колонками ['timestamp', 'open', 'high', 'low', 'close', 'volume'].
    - title: Заголовок графика.
    - window: Период для расчета Bollinger Bands и ATR (по умолчанию 20).
    - multiplier: Множитель для расчета Bollinger Bands (по умолчанию 2).
    """

    data = data.copy()

    # Рассчитываем Bollinger Bands
    data['sma'] = data['close'].rolling(window=window).mean()
    data['std'] = data['close'].rolling(window=window).std()
    data['upper_band'] = data['sma'] + (data['std'] * multiplier)
    data['lower_band'] = data['sma'] - (data['std'] * multiplier)

    # Рассчитываем ATR
    data['tr'] = pd.DataFrame({
        'high_low': data['high'] - data['low'],
        'high_close': abs(data['high'] - data['close'].shift()),
        'low_close': abs(data['low'] - data['close'].shift())
    }).max(axis=1)
    data['atr'] = data['tr'].rolling(window=window).mean()

    # Создаем график
    fig = make_subplots(rows=3, cols=1, shared_xaxes=True, vertical_spacing=0.1,
                        subplot_titles=(title, 'Volume', 'ATR'), row_width=[0.2, 0.2, 0.6])

    # График свечей с Bollinger Bands
    fig.add_trace(go.Candlestick(
        x=data['timestamp'],
        open=data['open'],
        high=data['high'],
        low=data['low'],
        close=data['close'],
        name='Candlesticks'
    ), row=1, col=1)

    # Bollinger Bands
    fig.add_trace(go.Scatter(
        x=data['timestamp'],
        y=data['upper_band'],
        line=dict(color='blue', width=1),
        name='Upper Band'
    ), row=1, col=1)

    fig.add_trace(go.Scatter(
        x=data['timestamp'],
        y=data['sma'],
        line=dict(color='orange', width=1),
        name='SMA'
    ), row=1, col=1)

    fig.add_trace(go.Scatter(
        x=data['timestamp'],
        y=data['lower_band'],
        line=dict(color='blue', width=1),
        name='Lower Band'
    ), row=1, col=1)

    # Столбчатый график объёмов
    fig.add_trace(go.Bar(
        x=data['timestamp'],
        y=data['volume'],
        name='Volume'
    ), row=2, col=1)

    # График ATR
    fig.add_trace(go.Scatter(
        x=data['timestamp'],
        y=data['atr'],
        line=dict(color='purple', width=1),
        name='ATR'
    ), row=3, col=1)

    # Настройки макета
    fig.update_layout(
        title=title,
        xaxis_title='Time',
        yaxis_title='Price',
        xaxis_rangeslider_visible=False
    )

    fig.show()
```

```python

plot_candlestick_with_indicators(zerobro_df, title='ZEREBROUSDT Candlestick with Bollinger Bands and ATR')
```

```python
plot_candlestick_with_indicators(ada_df, title='ADAUSDT Candlestick with Bollinger Bands and ATR')
```

```python
plot_candlestick_with_indicators(popcat_df, title='POPCATUSDT Candlestick with Bollinger Bands and ATR')
```
